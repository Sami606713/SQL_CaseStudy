**How many records are there in the dataset?**

```python
query="select count(*) as 'total_record' from exams"
pd.read_sql_query(query,conn)
```

**What is the average math score, reading score, and writing score?**

```python
query="""select avg(`math score`) over() as 'AVG_Math',
            avg(`reading score`) over() as 'AVG_Reading',
            avg(`writing score`) over() as 'AVG_writing'
            from exams
            limit 1
            """
pd.read_sql_query(query,conn)
```

**How many students took the test preparation course, and how many did not?**

```python
query="""select `test preparation course`,count(*) as 'Total'
        from exams
        group by `test preparation course`
        """
pd.read_sql_query(query,conn)
```

**How many female students are there in the dataset?**

```python
query="""select count(*) as 'Total Female'
            from exams
            where GENDER='female'
            """
pd.read_sql_query(query,conn)
```

**How many students belong to each race/ethnicity group?**

```python
query="""select `race/ethnicity`,count(*) as'each_group_student' 
            from exams
            group by `race/ethnicity`
            """
pd.read_sql_query(query,conn)
```

**How many students have a math score above 80?**

```python
query="""select count(*) as 'student>80' from exams
        where `math score`>80
            """
pd.read_sql_query(query,conn)
```

**How many students have a writing score below 70?**

```python
query="""select count(*) as 'student>80' from exams
        where `writing score`<70
            """
pd.read_sql_query(query,conn)
```

**Which gender has the highest average math score?**

```python
query="""select GENDER,avg(`math score`) as 'max_avg'
            from  exams
            group by GENDER
            order by max_avg desc
            limit 1
            """
pd.read_sql_query(query,conn)
```

**Which race/ethnicity group has the highest average reading score?**

```python
query="""select `race/ethnicity`,avg(`reading score`) as 'avg_reading'
        from exams
        group by `race/ethnicity`
        order by avg_reading desc
        limit 1
        """
pd.read_sql_query(query,conn)
```

**What is the maximum math score achieved?**

```python
query="""select max(`math score`) as 'max_math' from exams"""
pd.read_sql_query(query,conn)
```

**What is the minimum writing score achieved?**

```python
query="""select min(`writing score`) as 'max_writing' from exams"""
pd.read_sql_query(query,conn)
```

**Calculate the average math score, reading score, and writing score for each gender.**

```python
query="""select GENDER,avg(`math score`) as 'avg_math',
        avg(`writing score`) as 'avg_writing',
        avg('reading score') as 'avg_reading'
        from exams
        group by GENDER
        """
pd.read_sql_query(query,conn)
```

**Find the average scores for each race/ethnicity group.**

```python
query="""select `race/ethnicity`,avg(`math score`) as 'avg_math',
        avg(`writing score`) as 'avg_writing',
        avg('reading score') as 'avg_reading'
        from exams
        group by `race/ethnicity`
        order by `race/ethnicity` asc
        """
pd.read_sql_query(query,conn)
```

**Calculate the average scores for different parental levels of education.**

```python
query="""select `parental level of education`,avg(`math score`) as 'avg_math',
        avg(`writing score`) as 'avg_writing',
        avg('reading score') as 'avg_reading'
        from exams
        group by `parental level of education`
        """
pd.read_sql_query(query,conn)
```

**Find the students who scored above the average math score.**

```python
query="""select * from (select *,avg(`math score`)  over () as 'avg_math' 
            from exams)t
            where t.`math score`>t.avg_math
            """
pd.read_sql_query(query,conn)
```

**Find the students who belong to the race/ethnicity group A with the lowest average writing score.**

```python
query="""select * from(select * ,avg(`writing score`) over() as 'avg_writing'
        from exams 
        where `race/ethnicity`='group A') t
        where t.`writing score`<t.avg_writing
"""
pd.read_sql_query(query,conn)
```

**Rank the students based on their math score within each race/ethnicity group.**

```python
query="""select *, Rank() over(partition by `race/ethnicity` order by `math score` )as 'RANK' from exams"""
pd.read_sql_query(query,conn)
```
